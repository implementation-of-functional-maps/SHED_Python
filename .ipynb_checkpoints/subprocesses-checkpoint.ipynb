```python
# import subprocess
import shlex, subprocess
command_line = 'ls ls'
args = shlex.split(command_line)
args.append("cd ~/Desktop; ls")

print(args)
p = subprocess.Popen(args) # Success!
# cmd = "expect -c {set timeout 5; spawn env LANG=C /usr/bin/ssh katsurou@plata.ar.media.kyoto-u.ac.jp; expect \"password:\"; send \"aniyamato\n\"; }"
for cmd in args:
    output = subprocess.check_output( cmd, shell=True  ) 
    print(output)
    print("\n")
```
